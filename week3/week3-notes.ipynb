```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv("Height_Weight.csv")
```

```python
df.info()
```

```python
df[["Height","Weight"]].corr()
```

- Correlation is used to determine what features of a dataset could be used as an input

```python
data_input = df[["Height"]].values
data_output = df[["Weight"]].values
```

```python
data_input.shape
```

```python
plt.scatter(data_input, data_output)
```

```python
plt.hist(data_input,bins=50)
```

```python
plt.hist(data_output,bins=50)
```

```python
mm_scaler = MinMaxScaler()
input_norm = mm_scaler.fit_transform(data_input)
output_norm = mm_scaler.fit_transform(data_output)
```

```python
f, axes = plt.subplots(1, 2)
sns.histplot(input_norm, bins=50, ax=axes[0])
sns.histplot(output_norm, bins=50, ax=axes[1])
```

```python
std_scaler = StandardScaler()
input_norm = std_scaler.fit_transform(data_input)
output_norm = std_scaler.fit_transform(data_output)
```
